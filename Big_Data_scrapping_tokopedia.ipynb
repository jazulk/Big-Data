**Penjelasan value di toko**


---
Dalam data yang kita miliki pasti terdapat "value" adalah sebuah isi dari setiap variabel pada data, misal kita ambil dari data=

Pada kolom harga → value berupa angka harga seperti Rp1.500.000

Pada kolom rating → value berupa angka desimal seperti 4.9

Pada kolom diskon → value berupa persentase seperti 25%

Value ini muncul karena web scraping bekerja dengan cara menangkap isi dari setiap elemen yang dipilih (selector) di dalam struktur halaman web. Jadi, setiap informasi yang tampil di halaman toko akan disimpan sebagai value dalam dataset.
